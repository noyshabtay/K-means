```
from sklearn import datasets
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans

# import some data to play with
iris = datasets.load_iris()
X = iris.data[:, :]
y = iris.target


inertias = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=0)
    kmeans.fit(X)
    inertias.append(kmeans.inertia_)
plt.plot(range(1, 11), inertias)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.xlim(1, 10)
plt.ylabel('Inertia')
plt.annotate("Best K", xy=(3, 78.85144142614601),  xycoords='data',
            bbox=dict(boxstyle="round", fc="none", ec="gray"),
            xytext=(10, 30), textcoords='offset points', ha='center',
            arrowprops=dict(arrowstyle="->"))
plt.show()
print("Best K is 3 where the 'elbow' is. It makes sense since the dataset consists of 3 species of Iris.")
```
